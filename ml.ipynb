```python
df["Year"]=df["Year"].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["Make_Encode"]=lb.fit_transform(df["Make"])
```

```python
df["Model_Encode"]=lb.fit_transform(df["Model"])
```

```python
df["EFT_Encode"]=lb.fit_transform(df["Engine Fuel Type"])
```

```python
df["TransmissionType_Encode"]=lb.fit_transform(df["Transmission Type"])
```

```python
df["Driven_Wheels_Encode"]=lb.fit_transform(df["Driven_Wheels"])
```

```python
df["Vehicle Size_Encode"]=lb.fit_transform(df["Vehicle Size"])
```

```python
df["Vehicle Style_Encode"]=lb.fit_transform(df["Vehicle Style"])
```

```python
#Creating ML Models

X = df[["Year","Engine HP","Engine Cylinders","Number of Doors","highway MPG","city mpg","Popularity","Make_Encode","Model_Encode","EFT_Encode","TransmissionType_Encode","Driven_Wheels_Encode","Vehicle Size_Encode","Vehicle Style_Encode"]]
y = df['MSRP']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=49)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
```

```python
lr.fit(X_train,y_train)
```

```python
rf.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
rf.score(X_test,y_test)
```

```python
rf.score(X_train,y_train)
```
